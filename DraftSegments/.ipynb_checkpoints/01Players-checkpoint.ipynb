An attribute is nothing more than an ability that a player possesses and whose value is in the range of 0 to 99. The average rating of these attributes is reflected in the overall_rating field. The higher the value of these attributes, the better the player's ability. - We can classify the value of an attribute with the following quality labels. <BR>
	- Excellent: 90 - 99 <BR>
	- Very Good: 80 - 89 <BR>
	- Good: 70 - 79 <BR>
	- Fair: 50 - 69 <BR>
	- Poor: 40 - 49 <BR>
	- Very Poor: 0 - 39

OverallRating: Average rating of the player's skills <BR>
Potential: It is an estimate that indicates what the maximum overall_rating of a player will be during his professional career.<BR>
PreferredFoot: Indicates which leg the player tends to play with and therefore, the one on which he has greater power and control of the ball.<BR>
AttackWorkRate: The work rates indicate the level of involvement or participation of a player in attacking or defensive movements. These attributes can have the values high, medium and low. <BR>
DefenseWorkRate: The work rates indicate the level of involvement or participation of a player in attacking or defensive movements. These attributes can have the values high, medium and low. <BR>  
Crossing: Determines the quality and precision of a player making a pass towards the center of the field, near the opponent's goal. <BR> 
Finishing: Indicates the player's ability to score a goal.<BR> 
HeadingAccuracy: Sets the player's accuracy in passing or clearing the ball using their head.<BR> 
ShortPassing: Player's ability to make short passes <BR> 
Volleys: This skill demonstrates a player's ability to take shots or passes while the ball is in the air. <BR> 
Dribbling: Evaluates the ability to maintain possession of the ball against an opponent. This metric primarily measures the player's dribbling prowess. <BR> 
Curve: Establishes the player's ability to perform passes or shots with a curve effect.<BR> 
FreeKicAccuracy: Accuracy or skill of the player in free kicks.<BR> 
LongPassing: Player's ability to make long passes<BR> 
BallControl: Level of control of the player over the ball.<BR> 
Acceleration: Ability to increase the speed of the player on the field.<BR> 
SprintSpeed: Sets the speed of the player<BR> 
Agility: Evaluates how fast a player is able to handle the ball.<BR> 
Reactions: Response time of the player to events that happen around him.<BR> 
Balance: Shows the player's balance level.<BR> 
ShotPower: This attribute determines the power of a player's shots.<BR> 
Jumping: Rate the ability and quality of jumping.<BR> 
Stamina: Indicates the ability to maintain physical effort.<BR> 
Strength: Shows what the physical strength of a player is.<BR> 
LongShots: Ability of the player to take long shots.<BR> 
Aggression: Measures the frequency of pushing, pulling and blocking members of the opposing team.<BR> 
Interceptions: Ability to intercept the ball during passes from the opposing team.<BR> 
Positioning: Indicates the ability of the players to position themselves correctly on the field.<BR> 
Vision: Mental ability of the player to remember the positions of his teammates.<BR> 
Penalties: Accuracy of the player to take penalties.<BR> 
Marking: Ability to defend against a play by the opposing team<BR> 
StandingTackle: Ability of the player to make tackles, that is, to extend the leg to move the ball away from the opponent.<BR> 
SlidingTackle: Ability of the player to make tackles, that is, to extend the leg to move the ball away from the opponent
